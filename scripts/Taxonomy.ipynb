```python
'''
The function creates three additional columns to the BLAST-output-csv, where the genus, 
the phylum and the superkingdom are stored. If there is no entry in the lineage, the cell in the 
dataframe remains empty. 
'''
def df_tax(df):
    dfnew= df
    # store empty columns
    dfnew['genus']= ''
    dfnew['phylum']=''
    dfnew['superkingdom']=''
    
    
    for i in range(len(dfnew)):
            scinum = dfnew.staxids[i] # get tax id 
            lineage = ncbi.get_lineage(scinum) #get taxonomy of tax id (tax id numbers)
            
            ranks= ncbi.get_rank(i for i in lineage) # store the ranks of the lineage (genus, phylum, etc)
            
            ## storing the words (not tax ids) of the corresponding rank in the cells of the dataframe. 
            ## leave cell empfty if there is no rank given
            if 'genus' in list(ranks.values()):
                dfnew.genus[i] =list(ncbi.get_taxid_translator([str(list(ranks.keys())[list(ranks.values()).index('genus')])]).values())[0]
            if 'superkingdom' in list(ranks.values()):
                dfnew.superkingdom[i] = list(ncbi.get_taxid_translator([str(list(ranks.keys())[list(ranks.values()).index('superkingdom')])]).values())[0]
            if 'phylum' in list(ranks.values()):
                dfnew.phylum[i] = list(ncbi.get_taxid_translator([str(list(ranks.keys())[list(ranks.values()).index('phylum')])]).values())[0]
            
    return(dfnew)
```

```python
# read in csv (here: csv from html Cyano data)
workingcsv = pd.read_csv('BLAST_results.csv')
new_csv = df_tax(workingcsv)
```
